```python
#define time flitering function

def sst_tmfilt(sst,timesst,yearly,trend,seas,nym):
    
    #annual average sst (matlab)
    #
    #if yearly:
    #    timesst2=timesst(12:12:end);
    #    facm=1;
    #else
    #    timesst2=timesst;
    #    facm=12;
        
    #for python this part would be
    if yearly:
        timesst2=timesst[arange(12,int(timesst[-1])+1,12)];
        facm=1;
    if not(yearly):
        timesst2=timesst;
        facm=12;
    # here the size is assumed to give integer output. If not this could be a problem
    
    
    
    # remove trend (matlab)
    #--------------
    #tmp=sst*0;
    #for i=1:size(sst,1):
    #    for j=1:size(sst,2):
    #       if not(trend):
    #           tmp(i,j,:)=detrend(sst(i,j,:));
    #       if trend:
    #          tmp(i,j,:)=detrend(sst(i,j,:),'constant');
                
    #if not(trend):
    #    print('sst detrended')
    #    
    #
    #trnd=(sst-tmp);
    #sst=tmp;
    #sstmean=mean(trnd,3);
    #ssttrend=(trnd(:,:,end)-trnd(:,:,1))/(timesst2(end)-timesst2(1));
    
    # remove trend (python)
    #--------------
    # the detrend function is used as scipy.signal.detrend
    tmp=sst*0;
    for i in range(size(sst,0)):
        for j in range(size(sst,1)):
            if not(trend):
                tmp[i,j,:]=signal.detrend(sst[i,j,:]); #L fix
            if trend:
                tmp[i,j,:]=signal.detrend(sst[i,j,:],type='constant'); #L fix
                
                #here is a labelling matter
    
    if trend:
        print ('sst detrended in constant type')
            
        
    if not(trend):
        print('sst detrended')
        
    
    trnd=sst-tmp;
    sst=tmp;
    sstmean=mean(trnd[0]); #Error No.1. Not sure if fixed properlly
    ssttrend=(trnd[-1,:,:]-trnd[0,:,:])/(timesst2[0,-1]-timesst2[0,0]); #L fix
    
    #Remove seasonal cycle (matlab)
    #sstseas=sst(:,:,1:12)*0;
    #if not(yearly) and not(seas):
    #    for i=1:size(sst,1):
    #        for j=1:size(sst,2):
    #            for n=1:12:
    #                sstseas(i,j,n)=mean(sst(i,j,n:12:end));
    #                sst(i,j,n:12:end)=sst(i,j,n:12:end)-sstseas(i,j,n);
    #                
    #            
    #        
    #    print('seasonal cycle removed')
    
    #Remove seasonal cycle (Python)
    sstseas=sst[:,:,arange(12)]*0; #LL 
    
    print (shape(sstseas),'sstseasshape')
    
    if not(yearly) and not(seas):
        for i in range(size(sst,0)):
            for j in range(size(sst,1)):
                for n in range(12):
                    sstseas[i,j,n]=mean(sst[i,j,arange(n,size(sst,2),12)]);
                    sst[i,j,arange(n,size(sst,2),12)]=sst[i,j,arange(n,size(sst,2),12)]-sstseas[i,j,n];
                    
                
            
        print('seasonal cycle removed')
        
    
    #apply hanning window filter (matlab)
    #----------------------------
    #nmm=(nym)*facm/2;
    #
    #if nmm>0:
    #    coef=hanning(2*nmm-1)/nmm;
    #    #coef=ones(2*nmm-1,1)/(2*nmm-1);
    #    tmp=sst;
    #    for i=1:size(sst,1):
    #        for n=nmm:size(sst,3)-nmm+1:
    #            tmp(i,:,n)=squeeze(sst(i,:,n-nmm+1:n+nmm-1))*coef;
    #            
    #        
    #    sst=tmp(:,:,nmm:size(sst,3)-nmm+1);
    #    timesst2=timesst2(nmm:end-nmm+1);
    #
    #print('time hanning filter completed')
    
    #apply hanning window filter (python)
    nmm=(nym)*facm/2;
    
    if nmm>0:
        coef=hanning(2*nmm-1)/nmm; #numpy has a hanning function
        #coef=ones(2*nmm-1,1)/(2*nmm-1);
        tmp=sst;
        for i in range(0,size(sst,2)):
            for n in range(nmm,size(sst,0)-nmm+1):
                tmp[n,:,i]=squeeze(sst[arange(int(n-nmm+1),int(n+nmm,1)),:,i])*coef; #L fix
                
            
        sst=tmp[arange(int(nmm),int(size(sst,0)-nmm+2)),:,:]; #L fix
        timesst2=timesst2[arange(int(nmm),int(end-nmm+1))];

    print('time hanning filter completed')
    
    return sst,timesst2,facm,ssttrend,sstmean,sstseas
```

```python
#define eof function for eigenvalue analysis

def sst_eof(sst,xsst,ysst,latmin,latmax,lonmin,lonmax,neof):
    
    # indices of the spatial area (matlab)
    #indj=find(ysst>=latmin & ysst <=latmax);
    #if lonmax>lonmin
    #    indi=find(xsst<=lonmax & xsst>=lonmin);
    #else
    #    indi=find(xsst<=lonmax | xsst>=lonmin);
    #end
    
    #Equivalent python coding would be (further modification is applied)
    condj=zeros(size(ysst))
    indj=ysst*0
    for ii in arange(size(ysst)):
        condj[ii]=(ysst[ii]>=latmin and ysst[ii] <=latmax)
        if condj[ii]==True:
            indj[ii]=ii
    
    condi=zeros(size(xsst))
    indi=xsst*0
    if lonmax>lonmin:
        for ii in arange(size(xsst)):
            condi[ii]=(xsst[ii]<=lonmax and xsst[ii]>=lonmin)
            if condi[ii]==True:
                indi[ii]=ii
    if not (lonmax>lonmin):
        for ii in arange(size(xsst)):
            condi[ii]=(xsst[ii]<=lonmax or xsst[ii]>=lonmin)
            if condi[ii]==True:
                indi[ii]=ii
    
    indi=compress(condi,indi).astype(int) # to make them integers as indices
    indj=compress(condj,indj).astype(int)
    
    #here the brackets are changed from () to []. label 1-0 is not required here
    #xeof=xsst[indi];
    #yeof=ysst[indj];
    
    #use extract to get falues for eof from True/Falce index condition function
    
    xeof = extract(condi,xsst)
    yeof = extract(condj,ysst)
    
    # indices of the time span (matlab)
    #indsst=1:size(sst,3);
    #nsst=length(indsst);
    
    #equivalance in python would be
    indsst=arange(size(sst,0)) #L fix #made to be integer type
    nsst=size(indsst,0)
    
    # initialization of the matrice 2D space-time (matlab)
    #data=squeeze(sst(indi,indj,indsst));
    #A=size(data,1);
    #B=size(data,2);
    #data=permute(data,[3 1 2]);
    #data=reshape(data,nsst,[]);
    #size(data);
    
    #equivalent python lines would be
    #data=squeeze(sst[indsst,indj,indi]); #L fix
    sstsqueezing=sst[indsst,:,:]
    sstsqueezing=sstsqueezing[:,indj,:]
    sstsqueezing=sstsqueezing[:,:,indi]
    data=sstsqueezing
    #data=reshape(sstsqueezing,[1,size(sstsqueezing)]) #this is the data after the matlab squeeze function
    A=size(data,0);
    B=size(data,1);
    #data=expand_dims(data,axis=2)
    data=transpose(data,[2,0,1]); #L fix
    print shape(data)
    data=reshape(data,[nsst,-1])#,order="F");
    print shape(data); #not sure if this line matters
    #This part is highly non-confidential for the squeeze, permute and reshape functions
    #there is a labelling matter at the 3D matrices here.
    
    # eof
    #[pc,eig,eof]=svds(data(:,:),neof); the matlab linee
    #eof=reshape(eof,[length(xeof) length(yeof) neof]);
    
    #the equivalent phtyon lines are
    pcT, eig, eofT = ss.linalg.svds(data[:,:],neof) #the scipy.sparse.linalg.svds function is not the same as scipy.linalg.svd!
    #also notice that the eigenvalues are given as a 1Xn matrix instead of a nXn diagonal matrix, so the following line is needed:
    
    eig=diag(eig)
    # notice that eigenvalues are in increasing order for this function, which could be a problem if neof>1 and needs rearrangement.
    
    
    #eigindex=arange(1,int(neof+1))
    #eig=eigfull[eigindex]
    eof=eofT
    pc=pcT # be ware of consistency of the Transposes. The requirement for transpose is unknown for sparse.linalg.svds
    
    
    eof=reshape(eof,(neof, xeof.shape[0], yeof.shape[0])); #L fix #shape function dimension needs to be checked
    #labelling issue happen here
    
    eofreshape=transpose(eof,[0,2,1]) #get the right world map! Does not work yet!
    eofreshape=eofreshape[::-1,::-1,:] 
    
    #plt.figure(8)
    #plt.pcolor(eofreshape[0,:,:])
    #plt.colorbar()

    #for nplot in range(neof):
    #    plt.figure(nplot+5)
    #    plt.pcolor(eofreshape[nplot,:,:])
    #    plt.title('eigen function No. %3.0f'%(nplot+1))
    #    plt.colorbar()
    
    
    
    
    eof=eofreshape #ordreing in this data matrix would be [# of eigenvector,ydata,xdata]. Refer to this for future steps.
    
    # normalization of variance (matlab)
    #tracedata2=trace(data*transpose(data));
    #for n=1:neof
    #   var(n)=eig(n,n)**2/tracedata2;
    #end
   
    #normalization in python
    tracedata2=trace(dot(data,transpose(data))); #here matrix multiplication is used as dot(*,*)
    
    var=zeros(neof) #define var before assign its values
    print (tracedata2)
    for n in range(0,neof):
        var[n]=eig[n,n]**2/tracedata2;
    
    var=var[::-1] #correct the order of the variations
    # normalization of principal component (matlab)
    #for n=1:neof
    #  pc(:,n)=pc(:,n)/std(pc(:,n));
    #end
    
    # normalization of principal component in python
    for n in range(0,neof):
        pc[:,n]=pc[:,n]/std(pc[:,n], ddof=1); #L should work now
    #here is a labelling matter
    
    
    #The actual returning shape of pc and eof needs to be confirmed for consistency
    return eof,pc,var,xeof,yeof
```

```python
# this defines the regression function
# here some careful tests needs to be done to see if pc and eof are transposed in the right way.

def sst_regression(sst,pc,eof,xsst,ysst):
    
    #regression (matlab)
    #neof=size(pc,2);
    #reg=zeros(size(sst,1),size(sst,2),neof);
    #for n=1:neof
    #    for i=1:size(sst,1)
    #        for j=1:size(sst,2)
    #            reg(i,j,n)=pc(:,n)\squeeze(sst(i,j,:));
    #        end
    #    end
    #end
    
    #regression (python)
    neof=size(pc,1);
    reg=zeros([neof,size(sst,1),size(sst,0)]);
    
    print (shape(sst),'sstshape')
    print (shape(pc),'pcshape')
    
    #sst data is in the index form of [x,y,t]
    
    for n in range(0,neof):
        for i in range(0,size(sst,0)):
            for j in range(0,size(sst,1)):
                #pcregression=pc[:,n].reshape(-1,1) #we need 2D input at the first entry for solve function!
                #linalg.lstsq is used to give the same value as matlab backslash operator do for two column matrices
                reg[n,j,i]=linalg.lstsq(pc[:,n].reshape(-1,1),squeeze(sst[i,j,:]))[0]; #L not fixed for indices
            
            
    reg=transpose(reg,[0,2,1])  #This is to change reg in the format from [t,y,x] into [t,x,y] for plotting purpose
    reg=reg[::-1,:,:]  #This is to order the eigenvalues from largest to smallest
    print (shape(reg),'regshape')  

    #choix du signe dans une region donnee (matlab)
    #indys=find(ysst<=5 & ysst>=-5);
    #indxs=find(xsst<=280 & xsst>=250);
    #for n=1:neof
    #    if sum(sum(reg(indxs,indys,n)))<=0:
    #        pc(:,n)=-pc(:,n);
    #        reg(:,:,n)=-reg(:,:,n);
    #        eof(:,:,n)=-eof(:,:,n);
    #    end
    #end
    
    #choix du signe dans une region donnee (Python)
    #indys=where(ysst<=5 & ysst>=-5);
    #indxs=where(xsst<=280 & xsst>=250);
    
    #do it with extract function and multiple labels
    indj=ysst*0
    condj=indj
    for ii in arange(size(ysst)):
        condj[ii]=(ysst[ii]>=-5 and ysst[ii] <=5)
        if condj[ii]==True:
            indj[ii]=ii
    
    indi=xsst*0
    condi=indi
    for ii in arange(size(ysst)):
        condi[ii]=(xsst[ii]>=250 and xsst[ii] <=280)
        if condi[ii]==True:
            indi[ii]=ii
    
    indys=extract(condj,indj)
    indxs=extract(condi,indi)
    
    
    sumchecker=1
    if size(indxs)==0:
        sumchecker=0
        
    if size(indys)==0:
        sumchecker=0
    
    print sumchecker
    if sumchecker==1:
        for n in range(0,neof):
            if sum(sum(reg[n,indys,indxs]))<=0: #this is NOT the reg function in MATLAB!
                pc[n,:]=-pc[n,:]; #L fix
                reg[n,:,:]=-reg[n,:,:];
                eof[n,:,:]=-eof[n,:,:];
            
            
    return reg,pc,eof
```
